Q1. How can you create a Bokeh plot using Python code?

1. Install the Bokeh library using pip: pip install bokeh

2. Import the necessary modules from Bokeh: from bokeh.plotting import figure,  show, output_file

3. Define the data to be plotted (x and y values)

4. Create a new Bokeh plot using the figure() function and specifying the plot properties (e.g., plot width, plot height, x-axis label, y-axis label, etc.)
Add glyphs (markers, lines, etc.) to the plot using the circle(), line(), or other glyph functions.

5. Call the show() function to display the plot in the default browser or output_file() to save the plot to an HTML file.

```python
import numpy as np
from bokeh.plotting import figure, show, output_file

# Define data
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 5, 4, 6, 7])

# Create a new plot with the specified properties
p = figure(title="My Bokeh Plot", x_axis_label='X Axis', y_axis_label='Y Axis', plot_width=500, plot_height=400)

# Add glyphs to the plot
p.circle(x, y, size=10, color="navy", alpha=0.5)

# Show the plot in the default browser
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

```python
'''Glyphs in Bokeh are visual markers such as lines, circles, rectangles, and polygons that are used to display data on a plot. 
They are the basic building blocks of Bokeh plots, and they can be added to a plot using the figure method of the Bokeh library.'''

from bokeh.plotting import figure, show
import numpy as np

x = np.random.normal(size=100)
y = np.random.normal(size=100)
p = figure(title="Scatter plot", x_axis_label="X-axis", y_axis_label="Y-axis")
# Add circles as glyphs to the plot
p.circle(x, y, size=10, color='blue')
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
from bokeh.plotting import figure, show

# create a new plot with axis labels, title, and subtitle
p = figure(x_axis_label='X-axis', y_axis_label='Y-axis',
           title='My Plot',
           x_range=(0, 10), y_range=(0, 20))

# customize the legend
p.legend.location = 'top_left'
p.legend.orientation = 'vertical'

# customize the background
p.background_fill_color = 'lightgray'
p.border_fill_color = 'white'
p.outline_line_color = 'black'

# add some glyphs to the plot
p.circle([1, 2, 3, 4, 5], [2, 4, 6, 8, 10], size=10, color='blue', legend_label='Line 1')
p.line([1, 2, 3, 4, 5], [5, 7, 9, 11, 13], line_width=2, color='red', legend_label='Line 2')

# show the plot
show(p)
```
