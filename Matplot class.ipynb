```python
x=np.random.rand(50)
y=np.random.rand(50)
colours=np.random.rand(50)
size=1000*np.random.rand(50)
plt.scatter(x,y,c='#5BDF52',s=size,alpha=.5)
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('this is scatter graph b/t x and y')
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
colours=np.random.rand(50)
size=1000*np.random.rand(50)
plt.scatter(x,y,c='#5BDF52',s=size,alpha=.5,marker='x')
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('this is scatter graph b/t x and y')
```
