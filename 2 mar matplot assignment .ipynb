Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
# Answer:
#A scatter plot is a type of plot used to visualize the relationship between two variables.

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
# Answer:
# The subplot() function in Matplotlib is used to create multiple plots in the same figure, arranged in a grid.

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1) # Top left plot
plt.plot(x1, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2) # Top right plot
plt.plot(x2, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3) # Bottom left plot
plt.plot(x3, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4) # Bottom right plot
plt.plot(x4, y4)
plt.title("Line 4")

plt.tight_layout() # Optional: improves spacing between plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
# Answer:
'''A bar plot is a type of graph that represents categorical data with rectangular bars, 
    with the height or length of each bar proportional to the values it represents. 
    It is used to compare values across different categories.'''

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.subplot(2, 2, 1)
plt.bar(company, profit)
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")

plt.subplot(2, 2, 2)
plt.barh(company, profit)
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")

plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
# Answer:
'''A box plot is a graphical representation of the distribution of a dataset. It displays the range,median, quartiles, and outliers of the data. 
    It is used to identify the shape of the distribution, the presence of outliers, and to compare the distribution of two or more datasets.'''

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

# Plot box plot
plt.boxplot(data)
plt.xlabel('Data')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```
